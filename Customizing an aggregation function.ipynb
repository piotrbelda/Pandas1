```python
import pandas as pd
import numpy as np
```

```python
college=pd.read_csv('data/college.csv')
```

```python
college.head()
```

```python
college.groupby('STABBR')['UGDS'].agg(['sum','mean']).round(0).head()
```

```python
def max_deviation(col):
    std_score=(col-col.mean())/col.std()
    return std_score.abs().max()
```

```python
college.groupby('STABBR')['UGDS'].agg(max_deviation).round(1).head()
```

```python
college.groupby(['STABBR','RELAFFIL'])['UGDS','SATVRMID','SATMTMID'].agg([max_deviation,'mean','std']).round(1).head()
```
